## SENTIMENT ANALYSIS (POSITIVE / NEGATIVE ) IMDB MOVIE REVIEWS


### 1️⃣ Load Dataset

```python
import pandas as pd

# Load dataset
df = pd.read_csv("IMDB Dataset.csv")

# Taking a random sample of 3000 for faster processing
df = df.sample(3000, random_state=42)

# Display the first few rows
print(df.head())
```

### 2️⃣ Check Dataset Information

```python
# Get dataset summary information
print(df.info())

# Count number of positive and negative sentiments
print(df['sentiment'].value_counts())
```

### 3️⃣ Convert Sentiment Labels to Numeric Values

```python
# Convert 'positive' to 1 and 'negative' to 0
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})

# Verify transformation
print(df.head())  # This should now show numerical values for 'sentiment'
print(df['sentiment'].unique())  # Expected output: [1, 0]
```

### 4️⃣ Text Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert text into numerical format using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english', max_features=2000)

# Transform the 'review' column into numerical vectors
X = vectorizer.fit_transform(df['review']).toarray()
y = df['sentiment']

# Display shape of transformed data
print(X.shape)
```

### 5️⃣ Train-Test Split

```python
from sklearn.model_selection import train_test_split

# Split dataset into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Display shape of training and testing sets
print(X_train.shape, X_test.shape)
```

### 6️⃣ Measure Initial Training Accuracy

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Define different sample sizes to test initial accuracy
sample_sizes = [100, 500, 1000, 1500, 2000]
train_accuracies = []
train_losses = []

# Train a simple model on increasing sample sizes
for size in sample_sizes:
    X_sample, y_sample = X_train[:size], y_train[:size]
    model = RandomForestClassifier(n_estimators=10)
    model.fit(X_sample, y_sample)
    
    # Make predictions
    y_pred_sample = model.predict(X_sample)
    
    # Calculate accuracy and loss
    accuracy = accuracy_score(y_sample, y_pred_sample)
    loss = 1 - accuracy
    
    # Store results
    train_accuracies.append(accuracy)
    train_losses.append(loss)

# Print initial training accuracy values
print("Training Accuracy : ",train_accuracies)
print("Training Losses : ",train_losses)
```

### 7️⃣ Plot Training Accuracy and Loss Before Full Training

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))

# Plot training accuracy
plt.subplot(1, 2, 1)
plt.plot(sample_sizes, train_accuracies, marker='o', color='b', label='Training Accuracy')
plt.xlabel('Sample Size')
plt.ylabel('Accuracy')
plt.title('Training Accuracy Before Full Training')
plt.grid(True)

# Plot training loss
plt.subplot(1, 2, 2)
plt.plot(sample_sizes, train_losses, marker='o', color='r', label='Training Loss')
plt.xlabel('Sample Size')
plt.ylabel('Loss')
plt.title('Training Loss Before Full Training')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### 8️⃣ Initialize Classifiers

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier

# Define different classifiers for sentiment analysis
models = {
    'SVM': SVC(probability=True),
    'Random Forest': RandomForestClassifier(n_estimators=50),
    'Decision Tree': DecisionTreeClassifier(max_depth=5),
    'MLP': MLPClassifier(max_iter=500)  # Increase iterations from 100 to 500
}

```

### 9️⃣ Train Models and Evaluate Performance

```python
import numpy as np
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    roc_curve, auc, precision_recall_fscore_support
)

results = {}
after_training_accuracies = []
after_training_losses = []

# Train and evaluate each model
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else np.zeros(len(y_test))

    # Compute accuracy and loss
    acc = accuracy_score(y_test, y_pred)
    loss = 1 - acc
    after_training_accuracies.append(acc)
    after_training_losses.append(loss)

    # Compute precision, recall, and F1 score
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')
    
    # Compute confusion matrix and ROC curve
    cm = confusion_matrix(y_test, y_pred)
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    # Store results
    results[model_name] = {'Accuracy': acc, 'Precision': precision, 'Recall': recall, 'F1 Score': f1, 'AUC': roc_auc}
```

### 🔟 Visualizing Confusion Matrix for Each Model

```python
import seaborn as sns
import matplotlib.pyplot as plt

for model_name, model in models.items():
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{model_name} - Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

 ### 1️⃣1️⃣ Plotting ROC Curve for Each Model

```python
import matplotlib.pyplot as plt

for model_name, model in models.items():
    y_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else np.zeros(len(y_test))
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} - ROC Curve')
    plt.legend()
    plt.show()
```

 ### 1️⃣2️⃣ Comparing Training Accuracy and Loss After Full

```python
plt.figure(figsize=(10, 4))

# Bar plot for accuracy
plt.subplot(1, 2, 1)
plt.bar(models.keys(), after_training_accuracies, color='g')
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.title('Training Accuracy After Full Training')

# Bar plot for loss
plt.subplot(1, 2, 2)
plt.bar(models.keys(), after_training_losses, color='orange')
plt.xlabel('Classifier')
plt.ylabel('Loss')
plt.title('Training Loss After Full Training')

plt.tight_layout()
plt.show()
```

 ### 1️⃣3️⃣ Displaying Final Model Performance

```python
import pandas as pd

# Convert results into a DataFrame
result_df = pd.DataFrame(results).T

# Display final performance metrics
print(result_df)
```
